```python
import re
import duckdb
import polars as pl

MESSAGES_PARQUET_PATH = "dataset/LKML5Ws/*/*.parquet"
UNIQUE_AUTHORS_PATH = "dataset/unique_authors.parquet"
LKML_LIST = "org.kernel.vger.linux-kernel"

def extract_email(s: str | None) -> str | None:
    if not s:
        return None
    m = re.search(r"<([^>]+)>", s)
    if m:
        return m.group(1).strip().lower()
    # fallback: caso o campo seja só "email@dominio"
    s2 = s.strip().lower()
    return s2 if "@" in s2 and " " not in s2 else None

# ---------------------------
# A) Descobrir ids (hashes) do Greg e Linus na LKML via unique_authors
# ---------------------------
ua = pl.read_parquet(UNIQUE_AUTHORS_PATH)

lkml_authors = ua.filter(pl.col("list").list.contains(LKML_LIST)).with_columns(
    pl.col("__original_from").fill_null("").alias("__original_from"),
    pl.col("__original_from").str.to_lowercase().alias("from_raw_lc"),
    pl.col("__original_from").map_elements(extract_email, return_dtype=pl.Utf8).alias("email"),
)

# heurísticas iguais ao que você já viu no parquet
greg_authors = lkml_authors.filter(
    pl.col("from_raw_lc").str.contains("kroah") |
    pl.col("from_raw_lc").str.contains("greg kh") |
    pl.col("from_raw_lc").str.contains("greg kroah-hartman")
)

linus_authors = lkml_authors.filter(
    pl.col("from_raw_lc").str.contains("linus torvalds")
)

greg_ids = greg_authors.select(pl.col("from").unique()).to_series().to_list()
linus_ids = linus_authors.select(pl.col("from").unique()).to_series().to_list()

print("Greg IDs (LKML):", len(greg_ids))
print("Linus IDs (LKML):", len(linus_ids))

# mapa id -> email (alias) dentro da LKML
id_to_email = (
    pl.concat([
        greg_authors.with_columns(pl.lit("Greg Kroah-Hartman").alias("Sender")),
        linus_authors.with_columns(pl.lit("Linus Torvalds").alias("Sender")),
    ], how="vertical")
    .select(["Sender", "from", "email", "__original_from"])
    .filter(pl.col("email").is_not_null() & (pl.col("email") != ""))
    .unique()
)

# ---------------------------
# B) Contar mensagens por author_id no parquet grande (DuckDB)
# ---------------------------
con = duckdb.connect()

# conta quantas mensagens cada id enviou na LKML
counts = con.execute(f"""
    SELECT
        "from" AS author_id,
        COUNT(*) AS msg_count
    FROM read_parquet('{MESSAGES_PARQUET_PATH}')
    WHERE list = '{LKML_LIST}'
      AND "from" IS NOT NULL
    GROUP BY "from"
""").fetchdf()

counts_pl = pl.from_pandas(counts)

# manter só Greg/Linus
ids_all = set(greg_ids) | set(linus_ids)
counts_pl = counts_pl.filter(pl.col("author_id").is_in(list(ids_all)))

# ---------------------------
# C) Juntar: (Sender, email) + counts por author_id -> (Sender, Email, Count)
# ---------------------------
out = (
    id_to_email
    .rename({"from": "author_id"})
    .join(counts_pl, on="author_id", how="inner")
    .group_by(["Sender", "email"])
    .agg(pl.col("msg_count").sum().alias("Count"))
    .sort(["Sender", "Count"], descending=[False, True])
)

out.write_csv("email_alias_table_lkml.csv")
print(out)
print("Salvo em: email_alias_table_lkml.csv")
```

```python
import re
import duckdb
import polars as pl


MESSAGES_PARQUET_PATH = "dataset/LKML5Ws/*/*.parquet"
UNIQUE_AUTHORS_PATH = "dataset/unique_authors.parquet"
LKML_LIST = "org.kernel.vger.linux-kernel"

def extract_email(s: str | None) -> str | None:
    if not s:
        return None
    m = re.search(r"<([^>]+)>", s)
    if m:
        return m.group(1).strip().lower()
    s2 = s.strip().lower()
    return s2 if "@" in s2 and " " not in s2 else None


# -------------------------
# A) Descobrir IDs (hashes) do Greg e Linus na LKML
# -------------------------
ua = pl.read_parquet(UNIQUE_AUTHORS_PATH)

lkml_authors = ua.filter(pl.col("list").list.contains(LKML_LIST)).with_columns(
    pl.col("__original_from").fill_null("").alias("__original_from"),
    pl.col("__original_from").str.to_lowercase().alias("from_raw_lc"),
    pl.col("__original_from").map_elements(extract_email, return_dtype=pl.Utf8).alias("email"),
)

# Greg: baseado no que você já viu no parquet (kroah / greg kh / greg kroah-hartman)
greg_authors = lkml_authors.filter(
    pl.col("from_raw_lc").str.contains("kroah") |
    pl.col("from_raw_lc").str.contains("greg kh") |
    pl.col("from_raw_lc").str.contains("greg kroah-hartman")
)

# Linus
linus_authors = lkml_authors.filter(
    pl.col("from_raw_lc").str.contains("linus torvalds")
)

greg_ids = greg_authors.select(pl.col("from").unique()).to_series().to_list()
linus_ids = linus_authors.select(pl.col("from").unique()).to_series().to_list()

all_ids = sorted(set(greg_ids) | set(linus_ids))

print(f"Greg IDs (LKML): {len(greg_ids)}")
print(f"Linus IDs (LKML): {len(linus_ids)}")
print(f"Total IDs: {len(all_ids)}")

# salva os ids (auditoria)
pl.DataFrame(
    {"author_id": all_ids}
).write_csv("greg_linus_ids_lkml.csv")


# -------------------------
# B) Exportar TODAS as mensagens desses ids na LKML -> CSV
# -------------------------
con = duckdb.connect()

# monta uma lista SQL ('id1','id2',...)
ids_sql = ",".join([f"'{x}'" for x in all_ids])

query = f"""
    SELECT
        list,
        date,
        "from" AS author_id,
        subject,
        "message-id" AS message_id,
        "in-reply-to" AS in_reply_to,
        "references" AS references,
        raw_body
    FROM read_parquet('{MESSAGES_PARQUET_PATH}')
    WHERE list = '{LKML_LIST}'
      AND "from" IN ({ids_sql})
    ORDER BY date ASC
"""

df = con.execute(query).fetchdf()

# Exporta CSV
out_csv = "lkml_messages_greg_linus.csv"
df.to_csv(out_csv, index=False)

print(f"Mensagens exportadas: {len(df)}")
print(f"CSV salvo em: {out_csv}")
```
